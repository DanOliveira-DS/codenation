# Dados Categoricos
Este tipo de dado representa as carateristicas de um objeto, por exemplo, **sexo, estado civil, sim e não**.. etc. Fazem parte do conjunto de dados **qualitativos**. Podem ser Binários (variável **dicotómica**) ou variáveis **politómicas** 

# Escala de medição
Existem quatro escalas de medição variáveis: **nominal, ordinal, intervalo e relação**. Estas escalas de medição são formas de classificar diferentes variáveis. Por padrão, todas as variáveis se enquadram numa das quatro escalas acima mencionadas. Como dito antes

A maior parte do conjunto de dados categóricos segue escalas nominais ou ordinais.

## Nominal
São utilizadas para rotular variáveis sem qualquer valor quantitativo. Não tem qualquer importância numérica.
A escala nominal possui apenas a característica de descrição, o que significa que possui rótulos únicos para identificar ou delegar valores aos itens. 
* Exemplos: 0 – Masculino e 1 – Feminino, número na camiseta dos jogadores e nos carros de corrida

## Ordinal
A escala ordinal é o segundo nível de medição que reporta a classificação e ordenação dos dados sem realmente estabelecer o grau de variação entre eles. Os dados ordinais são dados quantitativos que têm ordens naturais e a diferença entre eles é desconhecida. Podem ser nomeados, agrupados e também classificados.

Exemplos: pesquisas de forma em geral
* Você está satisfeito com o atendimento ao cliente?
    1. Muito infeliz
    2. infeliz
    3. Neutro
    4. Infeliz
    5. Muito infeliz


## Intervalo (*Range*)
Nas escalas de intervalo, tanto a ordem como as diferenças exatas entre os valores são significativas. As escalas de intervalo são amplamente utilizadas na estatística, por exemplo nas medidas das tendências centrais (media, mediana, moda) e desvio padrão.

Exemplos:
* Os dados de intervalo podem ser discretos com números inteiros, como 8 graus, 4 anos, 2 meses, etc., ou contínuos com números fracionários, como 12,2 graus, 3,5 semanas ou 4,2 milhas.



## Relação (*Ratio*)
As escalas de relação contem ordem, valores exatos e zero absoluto, o que torna possível a sua utilização em estatísticas descritivas e inferenciais. Estas escalas oferecem inúmeras possibilidades de análise estatística. As operações matemáticas, a medida das tendências centrais e a medida de dispersão e coeficientes de variação também podem ser calculadas a partir dessas escalas.

A grande diferença entre a escala de intervalo e relação fica por conta do zero absoluto. A escala de relação não possui valores negativos.


# Estatistica Descritiva
Como o nome sugere, ajudam a descrever e compreender os conjuntos de dados, apresentando-os de uma maneira mais significativa o que permite uma interpretação mais simples dos dados. Não é usada para tirar conclusões além dos dados analisados ou para chegar a conclusões sobre quaisquer hipóteses feitas. Os tipos mais comuns de estatísticas descritivas incluem a medida de tendências centrais, a medida de desvio e outras.


## Medidas de tendência central
Representa o ponto central ou valor típico de um conjunto de dados. Estas medidas indicam onde “cai” a maioria dos valores de uma distribuição e são também referidas como a localização central de uma distribuição. As tês medidas mais comuns de tendência central são a **média**, a **mediana** e a **moda**. Cada uma delas calcula a localização do ponto central utilizando um método diferente.

obs.: A escolha da melhor medida de tendência central depende do tipo de dados que se dispõe

### Média (*mean*)
É um número em torno do qual as variáveis contínuas observadas estão distribuídas. Este número estima o valor da totalidade do conjunto de dados. Matematicamente, é o resultado da divisão da soma dos números pelo número de números inteiros do conjunto de dados.

$$\mu= \frac{1}{n} \sum_{i=1}^{n} xi$$

**obs.: A média é facilmente influenciada por valores extremos (*outliers*). Isso ocorre em distribuições enviesadas (*skewed*) onde a média é puxada para longe do centro. Consequentemente, é melhor usar a média como medida de tendência central quando se tem uma distribuição simétrica (gaussiana)**


### Mediana (*median*)
É calculada ordenando primeiro todos os dados e depois localizando o valor médio da amostra. Para observações com um numero par, a mediana é calculada como a média das duas observações do meio.

**obs.: Ao contrário da média. Distribuições enviesadas e outliers pouco afetam a mediana. É uma medida indicada para distribuições enviesadas, dados contínuos e ordinais.**

Vamos representar essas diferenças de forma gráfica. Utilizarei o conjunto de dados 'pima-indians-diabetes.csv' que pode ser encontrado no [kaggle](https://www.kaggle.com/uciml/pima-indians-diabetes-database)

```python
# importando dataset 
pima = pd.read_csv('pima-indians-diabetes.csv', header=None)

# criando figura 2 linhas e 2 colunas
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
max_xlim, max_ylim = plt.ylim()

# Distribuicao simetrica - media
pima[5].hist(grid=False, ax=ax[0][0])
ax[0][0].axvline(pima[5].mean(), color='k', linestyle='-', linewidth=2)
ax[0][0].text(pima[5].mean()*1.005, max_ylim+290, 'Média: {:.2f}'.format(pima[5].mean()), fontsize=12)

# Distribuicao simetrica - mediana
pima[5].hist(grid=False, ax=ax[0][1])
ax[0][1].axvline(pima[5].median(), color='r', linestyle='-', linewidth=2)
ax[0][1].text(pima[5].median()*1.005, max_ylim+290, 'Mediana: {:.2f}'.format(pima[5].median()), color='red', fontsize=12)

#distribuicao enviesada - media
pima[0].hist(grid=False, ax=ax[1][0])
ax[1][0].axvline(pima[0].mean(), color='k', linestyle='-', linewidth=2)
ax[1][0].text(pima[0].mean()*1.005, max_ylim+260, 'Média: {:.2f}'.format(pima[0].mean()), fontsize=12)

#distribuicao enviesada - mediana
pima[0].hist(grid=False, ax=ax[1][1])
ax[1][1].axvline(pima[0].median(), color='r', linestyle='-', linewidth=2)
ax[1][1].text(pima[0].median()*1.005, max_ylim+260, 'Mediana: {:.2f}'.format(pima[0].median()), color='red', fontsize=12)

plt.tight_layout(pad=1)
```

Os gráficos reforçaram a ideia de que em distribuições simétricas tanto a media e mediana fixaram no mesmo ponto. Já em uma distribuição enviesada foi possível ver a disparidade entre as duas medidas

### Moda
É o número inteiro que aprece no conjunto de dados no número máximo de vezes. Valor com maior frequência no conjunto de dados.

**obs.: É uma medida de tendência central para dados categóricos, ordinal e distribuição de probabilidade**

Para nosso conjunto de dados da dengue a moda para `casos-confirmados` é 0

## Medidas de Dispersão
Representa a quantidade de dispersão num conjunto de dados. Enquanto uma medida de tendência central descreve o valor típico, as medidas de variabilidade definem a distância a que os pontos de dados tendem a '‘cair’' do centro. Uma baixa dispersão indica que os pontos de dados tendem a ser agrupados fortemente em torno do centro. Uma dispersão elevada significa que eles tendem a cair mais longe. As medidas de variabilidade são **range**, **amplitude interquartil**, **variância**, **desvio padrão**, **enviesamento** e **curtoses**.

### Range
É a medida de variabilidade mais simples de calcular e a mais simples de compreender. O intervalo (*range*) de um conjunto de dados é a diferença entre o maior e o menor dos valores desse conjunto de dados. Como é baseado em valores extremos são suscetíveis a *outliers*

```python
# exemplo
max_value = dengue['casos-confirmados'].max()
min_value = dengue['casos-confirmados'].min()
range_dengue = max_value - min_value
print(range_dengue)
```

### Amplitude Interquartil (IQR)
É uma medida robusta de variabilidade de forma semelhante a mediana, que é uma medida robusta de tendência central. Nenhuma das medidas é influenciada dramaticamente por *outliers* porque não dependem de todos os valores. Adicionalmente, o intervalo interquartil é excelente para distribuições enviesadas.

```python
PATH = "/home/dscoliveira/Pictures/"
Image(filename = PATH + "iqr.png", width=500, height=100)
```

```python
# calculo do IQR
casos = dengue['casos-confirmados'].value_counts()
q1 = np.percentile(casos, 25)
q2 = np.percentile(casos, 50)
q3 = np.percentile(casos, 75)

IQR = q3 - q1
print(IQR)
```

```python
casos.plot(kind='box', vert=False, figsize=(10, 8))
plt.tight_layout()
```

### Variância
É a diferença média quadrática dos valores da média. Ao contrário das medidas anteriores de variabilidade, a variância inclui todos os valores no calculo através da comparação de cada valor com a média. Uma variância baixa tera valores agrupados em torno da media, enquanto uma variância alta tera valores dispersos a partir da media

Temos duas formas de calcular a variância, para amostra ou população:
* Amostra
$$s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(xi - \mu)^2$$

* população
$$\sigma^2=\frac{1}{n}\sum_{i=1}^{n}(xi - \mu)^2$$

```python
# calculando variancia utilizando o funcao var() do pandas
variance = dengue.var()
print(f"{variance}\n")

# variance de uma coluna especifica
var_casos = dengue['casos-confirmados'].var()
print(f" Variância para coluna de casos confirmados é de: {var_casos}")
```

```python
dengue['casos-confirmados'].mean()
```

### Desvio Padrão
É difícil interpretar a variação porque as unidades são as unidades quadradas das observações. Podemos retornar as unidades às unidades originais das observações, tomando a raiz quadrada do resultado. Isso é chamado de desvio padrão. O desvio padrão, juntamente com a média, são os dois parâmetros principais necessários para especificar qualquer distribuição gaussiana. Assim como a variância temos desvio padrão para amostra e população

* Amostra
$$s = \sqrt{\frac{1}{n-1}\sum_{i=1}^{n}(xi - \mu)^2}$$

* População
$$\sigma=\sqrt{\frac{1}{n}\sum_{i=1}^{n}(xi - \mu)^2}$$

```python
# calculando desvio padrao utilizando o funcao std() do pandas
des_padrao = dengue.std()
print(f"{des_padrao}\n")

# variance de uma coluna especifica
std_casos = dengue['casos-confirmados'].std()
print(f" Desvio padrao para coluna de casos confirmados é de: {std_casos}")
```

### Curtoses (*kurtosis*)
Ilustra o quanto as caudas da distribuição diferem da de uma distribuição normal. Pode identificar se uma dada distribuição contem *outliers*

* Mesocúrtica: Possui curtose em torno de 0.
* Leptocúrtica: A distribuição tem curtose maior que 3 e as caudas indicam que a distribuição produz mais discrepâncias (*outliers*).
* Platicúrtica: A distribuição tem curtose negativa e as caudas são mais finas em comparação a distribuição normal.

```python
# calculando curtoses utilizando função kurt()
kurtosis = dengue.kurt()
print(kurtosis)
```
